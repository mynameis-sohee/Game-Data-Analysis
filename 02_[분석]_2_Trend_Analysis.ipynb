```python
# Grouping
df['Year_GR'] = df['Year'].astype('string').str[:3].astype('int')
df['Year_GR'].astype('object')
df['Sales'] = df['NA_Sales']+df['EU_Sales']+df['JP_Sales']+df['Other_Sales']
df_trend_group = df.groupby(['Year_GR','Genre']).Sales.sum().reset_index().sort_values(by='Sales', ascending=False).reset_index(drop=True)
df_trend_group.head(3)
```

```python
# There are two independent variables -> Use 2-way ANOVA
# There is a statistically significant trend ! (C(Genre):C(Year_GR))

model = ols('Sales ~ C(Genre)* C(Year_GR)', df).fit()
anova_lm(model)
```
